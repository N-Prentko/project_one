```python
# testing getting values from the FEC api
purpose_url = 'https://api.open.fec.gov/v1/schedules/schedule_b/by_purpose/'

# set the amount of data to be returned per page. Must be under 100
per_page = 50

# params to pass into the url
params = {'api_key' : fec_key,
         'per_page' : per_page}

# first json request to get the number of pages to loop through
purpose_search = requests.get(purpose_url, params=params)

# json
purpose_search = purpose_search.json()

# store the number of pages the search returns
page_nums = purpose_search['pagination']['pages']

print(json.dumps(purpose_search, indent=4))
```
